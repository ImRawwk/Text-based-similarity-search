### Performing Text Lemmetization

```python
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
import pandas as pd

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Define the lemmatization function
def lemmatize_text(text):
    lemmatized_words = []
    words = nltk.word_tokenize(text)

    for word in words:
        pos_tag = nltk.pos_tag([word])[0][1][0].upper()
        pos_tag = pos_tag if pos_tag in ['N', 'V', 'R'] else wordnet.NOUN
        lemmatized_word = lemmatizer.lemmatize(word, pos=pos_tag.lower())
        lemmatized_words.append(lemmatized_word)

    lemmatized_text = " ".join(lemmatized_words)
    return lemmatized_text
```

### Preprocessing data 

```python
def preprocess_text(text):
    """
    Takes in raw text in string format.
    """
    # returns cleaned text
    # lowercase the text
    text = text.str.lower()
    # remove numbers & special characters
    text = re.sub('[^A-Za-z0-9]+', ' ', text)
    # remove stopwords
    text = ' '.join(e for e in text.split() if e not in stopwords)
    # perform text lemmetization
    text = lemmatize_text(text)
    # replace double space with single space
    text = text.replace("  ", " ")
```

```python
# Importing all packages required for FAISS
import faiss
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
# Loading sentence transformer model from huggingface 
model = SentenceTransformer('sente nce-transformers/all-mpnet-base-v2')
```

```python
# Loading original dataframe
import pandas as pd
df = pd.read_csv('data.csv')
df.head()
```

```python
### Embedding the text descriptions from original dataframes
sentences = list(df['Product description'])
embeddings = model.encode(sentences)
```

```python
## adding faiss index for efficient searching
index = faiss.IndexFlatIP(embeddings.shape[1])  # Initialize the index
index.add(embeddings)  # Add the embeddings to the index
```

```python
# Defining the query text & embedding it
query = 'ASOS DESIGN t-shirt with crew neck and roll sleeve in white'
query_embedding = model.encode([query])  # Encode the query sentence

# Perform the similarity search
D, I = index.search(np.array(query_embedding), k=5)  # Retrieve top 5 most similar sentences

# Display the results
for i, idx in enumerate(I[0]):
    print(f"Similarity Rank {i+1}: {sentences[idx]}")
```
